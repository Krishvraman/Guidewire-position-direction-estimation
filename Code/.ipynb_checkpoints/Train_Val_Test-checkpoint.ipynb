### Milestone 1 - Study the Data Structure

```python
# Analysing the given dataset 
fdataset = pathlib.Path(drive +'GuidewireDataset.npz')
with np.load(fdataset) as f:
    dataset = dict(f)
```

```python
# Analyzing the data structre 
print(dataset.keys())
print(dataset['images'].shape)
print(dataset['directions'].shape)
print(dataset['positions'].shape)
Number_of_Images=len(dataset['images'])
print(Number_of_Images)
```

### Milestone 2  - Preprocess and plot images with guidewire tips as bounding boxes

```python
# Looping through each image, position and direction - from the dictionary - Doing some processing to plot 
images = [x.astype(float) for x in dataset['images']] # Making images in floating point dtype
images2 = [exposure.equalize_hist(x) for x in images] # Performing histogram equalization
images = [np.interp(x, (0,1), (np.min(x),np.max(x))) for x in images2] # Making the range from 0 to 1 
images = [1 - x for x in images] # Making the guidewire(target) - to have the higher side of contrast - to boost the network 
positions = [np.transpose(x) for x in dataset['positions']] # Extracting the position - transposing to get x | y in columns than in rows - to plot easily 
directions = [np.transpose(x) for x in dataset['directions']]# Extracting the direction  transposing to get x | y in columns than in rows - to plot easily 
```

```python
# Function to plot an image - given index 
def plot_image(id,window_size_half):
  i = id  # image index
  w = window_size_half  # window size
  for position, direction in zip(positions[i].T, directions[i].T):
      j1, j2 = round(position[1] - w), round(position[1] + w + 1)
      k1, k2 = round(position[0] - w), round(position[0] + w + 1)
      # randomly move y coordinate of the center
      # dj = round(np.random.uniform(-10, 10))
      # j1 += dj
      # j2 += dj
      # randomly move x coordinate of the center
      # dk = round(np.random.uniform(-10, 10))
      # k1 += dk
      # k2 += dk
      #print(j1,j2,k1,k2)
      fig, ax = plt.subplots(figsize=(3,3))
      ax.imshow(images2[i], cmap='gray')
      ax.add_patch(plt.Rectangle((k1, j1), k2 - k1 + 1, j2 - j1 + 1, ec='magenta', fc='none'))
      #ax.imshow(images[i][j1:j2,k1:k2], cmap='gray')
      center = position[0],position[1]
      ax.plot(*center, 'o', mfc='none', color='yellow')
      ax.quiver(*center, *direction, angles='xy', color='yellow')
      angle = np.rad2deg(np.arctan2(*direction))
      if angle < 0:
          angle = 360 + angle
      ax.set_title(f'{angle:.1f} deg')
```

```python
#Plotting first 5 x2 images 
for n in range(20):
   plot_image(n,50)
```

### MileStone 3 - Create Train - Test - Val split

```python
list_indices=[i for i in range(Number_of_Images)]
print(list_indices)
```

```python
# Train Val and Test split - 70% 10% 20% split
in_trainval, in_test = train_test_split(list_indices, test_size=0.20,shuffle=True ,random_state=3)
in_train, in_val = train_test_split(in_trainval, test_size=0.08,shuffle=True,random_state=3)
print("Train Images Indices: ",in_train)
print("\n\nValidation Images Indices: ",in_val)
print("\n\nTest Images Indices: ",in_test)
```

### MileStone 4 : Save the indices to a file - to be used in the other model building codes

```python
in_train_arr=np.array(in_train)
in_val_arr=np.array(in_val)
in_test_arr=np.array(in_test)
print(in_val_arr)
```

```python
np.save(drive+"Train_Indices.npy",in_train_arr)
np.save(drive+"Val_Indices.npy",in_val_arr)
np.save(drive+"Test_Indices.npy",in_test_arr)
```
