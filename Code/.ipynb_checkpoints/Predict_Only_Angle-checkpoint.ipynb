### Milestone 1 - Understanding dataset - Preprocessing 

```python
# Analysing the given dataset 
fdataset = pathlib.Path(drive +'GuidewireDataset.npz')
with np.load(fdataset) as f:
    dataset = dict(f)
```

```python
# Analyzing the data structre 
print(dataset.keys())
print(dataset['images'].shape)
print(dataset['directions'].shape)
print(dataset['positions'].shape)
```

```python
# Looping through each image, position and direction - from the dictionary
images = [x.astype(float) for x in dataset['images']] # Making images in floating point dtype
images = [exposure.equalize_hist(x) for x in images] # Performing histogram equalization
images = [np.interp(x, (0,1), (np.min(x),np.max(x))) for x in images] # Making the range from 0 to 1 
images = [1 - x for x in images] # Making the guidewire(target) - to have the higher side of contrast - to boost the network 
positions = [np.transpose(x) for x in dataset['positions']] # Extracting the position - transposing to get x | y in columns than in rows - to plot easily 
directions = [np.transpose(x) for x in dataset['directions']]# Extracting the direction  transposing to get x | y in columns than in rows - to plot easily 
```

```python
# Function to plot an image - given index 
def plot_image(id,window_size_half):
  i = id  # image index
  w = window_size_half  # window size
  for position, direction in zip(positions[i].T, directions[i].T):
      j1, j2 = round(position[1] - w), round(position[1] + w + 1)
      k1, k2 = round(position[0] - w), round(position[0] + w + 1)
      # randomly move y coordinate of the center
      # dj = round(np.random.uniform(-10, 10))
      # j1 += dj
      # j2 += dj
      # randomly move x coordinate of the center
      # dk = round(np.random.uniform(-10, 10))
      # k1 += dk
      # k2 += dk
      fig, ax = plt.subplots(figsize=(3,3))
      ax.imshow(images[i][j1:j2,k1:k2], cmap='gray')
      center = (j2 - j1 - 1) / 2, (k2 - k1 - 1) / 2
      ax.plot(*center, 'o', mfc='none', color='yellow')
      ax.quiver(*center, *direction, angles='xy', color='yellow')
      angle = np.rad2deg(np.arctan2(*direction))
      if angle < 0:
          angle = 360 + angle
      ax.set_title(f'{angle:.1f} deg')
```

```python
#Plotting first 5 x2 images 
for n in range(5):
   plot_image(n,50)
```

```python
# Generating the dataset
Generated_images=[]
Angles=[]
number_of_images=len(images)
for id in range(number_of_images):
  i = id  # image index
  w = 50  # window size - half (since left and right)
  for position, direction in zip(positions[i].T, directions[i].T):
      j1, j2 = round(position[1] - w), round(position[1] + w + 1)
      k1, k2 = round(position[0] - w), round(position[0] + w + 1)
      # randomly move y coordinate of the center
      # dj = round(np.random.uniform(-10, 10))
      # j1 += dj
      # j2 += dj
      # randomly move x coordinate of the center
      # dk = round(np.random.uniform(-10, 10))
      # k1 += dk
      # k2 += dk
      generated_image=images[i][j1:j2,k1:k2]
      angle = np.rad2deg(np.arctan2(*direction))
      if angle < 0:
          angle = 360 + angle
      
      Generated_images.append(generated_image)
      Angles.append(angle)
```

```python
print(len(Generated_images)) # Should be 314 x 2 = 628
print(len(Angles)) # Should be 314 x 2 = 628
```

```python
# Making dataset - pytorch form - so that we can control things later if we want to 
class Dataset_Centered_Guide_Wire(torch.utils.data.Dataset):
  def __init__(self, Gen_im_list, angles_list, transform=None):
    super().__init__()
    self._transform = transform
    self._angles=angles_list
    self._images=Gen_im_list

  def __len__(self):
    return len(self._images)

  def __getitem__(self, index):
    image=self._images[index]
    image=np.expand_dims(image, axis=0)
    image = image.astype(np.float32)  
    image = torch.as_tensor(image.copy())  # cast NumPy array to Torch tensor
    label=self._angles[index]
    label = torch.as_tensor(label.copy())
    return image,label
```

## Milestone 2 - Making train-val-test split

```python
# Converting list to array
Gen_array = np.array(Generated_images)
angle_array=np.array(Angles)
```

```python
# Train Val and Test split - 70% 10% 20% split
im_trainval, im_test, ang_trainval, ang_test = train_test_split(Gen_array, angle_array, test_size=0.20,shuffle=True ,random_state=3)
im_train, im_val, ang_train, ang_val = train_test_split(im_trainval, ang_trainval, test_size=0.08,shuffle=True,random_state=3)
```

```python
# Checking size of Train - val - Test
print(im_train.shape)
print(ang_train.shape)

print(im_val.shape)
print(ang_val.shape)

print(im_test.shape)
print(ang_test.shape)
```

```python
# Making the Dataset loaders
Train_loader= torch.utils.data.DataLoader(dataset=Dataset_Centered_Guide_Wire(im_train,ang_train,transform=None),batch_size=10)
Val_loader= torch.utils.data.DataLoader(dataset=Dataset_Centered_Guide_Wire(im_val,ang_val,transform=None),batch_size=10)
Test_loader= torch.utils.data.DataLoader(dataset=Dataset_Centered_Guide_Wire(im_test,ang_test,transform=None))
```

```python
# Plotting samples from our train dataloader 
nimages = 5
i=0
for images, labels in Train_loader:
  for image, label in zip(images, labels):
    if i==nimages:
      break
    i=i+1
    plt.imshow(np.squeeze(image,axis=0), cmap='gray')
    plt.title('Angle is '+str(round(label.item(),2)))
    plt.show()
```

### Milestone 3 - Building the model

```python
# Creating a loss function - Idea is to get cos and sin output from network and calculate loss
def Sine_Cos_MSE_loss(sine_out,cosine_out,GT_angle_degree):
  #All calculations in radians 
  Original_Sine=torch.sin(torch.deg2rad(GT_angle_degree))
  Original_Cosine=torch.cos(torch.deg2rad(GT_angle_degree))
  #The loss Function
  MSE_loss=torch.mean(((sine_out-Original_Sine)**2)+((cosine_out-Original_Cosine)**2))

  return MSE_loss
```

```python
# Building a model for regression  - predicting the angle
resnet = models.resnet50(pretrained=True)
print(resnet)
```

```python
resnet.conv1 = torch.nn.Conv2d(1, 64, (7, 7), (2, 2), (3, 3), bias=False)
resnet.fc=torch.nn.Sequential(torch.nn.Linear(2048, 2),torch.nn.Tanh())
```

```python
print(resnet)
```

```python
if torch.cuda.is_available():
    print("GPU Available")
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")  
model = resnet.to(device)
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3, momentum=0.9) 
cost = Sine_Cos_MSE_loss
```

```python
num_epochs = 500  # number of epochs to training for
losses_train = []  # keep track of loss per epoch
losses_val = []

for epoch in tqdm(range(num_epochs)):
    model.train()  # switch to training mode

    loss_train = 0
    for images, labels in Train_loader:
        optimizer.zero_grad()  # zero optimizer gradients, default option accummulates!
        outputs = model(images.to(device))  # run the forward model   
        Sine_output=outputs[:,0]
        cosine_output=outputs[:,1]     
        loss = cost(Sine_output,cosine_output,labels.to(device))  # compute loss
        loss.backward()  # backpropagate loss
        optimizer.step()  # step the optimizer along the gradient

        # keep track of training loss
        loss_train += loss.item()
    loss_train /= Train_loader.batch_size  # optional, normalize loss by batch size
    losses_train.append(loss_train)

    # compute and keep track of validation loss
    with torch.no_grad():
        loss_val = 0
        for images, labels in Val_loader:
            outputs = model(images.to(device))
            Sine_output=outputs[:,0]
            cosine_output=outputs[:,1]     
            loss = cost(Sine_output,cosine_output,labels.to(device))  # compute loss
            loss_val += loss.item()
        loss_val /= Val_loader.batch_size
        losses_val.append(loss_val)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
lines1 = ax.plot(losses_train, 'C0')
ax.set_ylabel('Train Loss')
axr = ax.twinx()  # show validation loss on the right axes
lines2 = axr.plot(losses_val, 'C1')
axr.set_ylabel('Val Loss')
ax.set_xlabel('Epoch')
ax.legend([lines1[0], lines2[0]], ['train', 'val'])
pass
```

```python
# Predict and Evaluate
labels_l = []
predictions_l =[]
with torch.no_grad():
    model.eval()
    for images, labels in tqdm(Test_loader):
        outputs = model(images.to(device))
        sine_predictions = outputs[:,0].cpu()
        cosine_predictions = outputs[:,1].cpu()
        predicted_angle = np.rad2deg(np.arctan2(sine_predictions.item(),cosine_predictions.item()))
        if predicted_angle<0:
            predicted_angle+=360
        labels_l.append(labels.item())
        predictions_l.append(predicted_angle)
```

```python
print(labels_l[0:10])
print(predictions_l[0:10])
```

```python
# Making sure our error is taking the fact that error above 180 degree is less
# Example : error between 359 and 1 is actually 2 degrees and not 358 degree
Errors=abs(np.subtract(labels_l,predictions_l))
print(Errors[0:10])
updated_error=[360-er if er>180 else er for er in Errors]
print(updated_error[0:10])
```

```python
#Getting mean absolute error
MAE=sum(updated_error)/len(updated_error)
```
